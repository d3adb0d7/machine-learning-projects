```python
from sklearn.model_selection import train_test_split
```

```python
new_data_set = pd.read_csv('data/cardio_train.csv',sep=';')
```

```python
new_data_set
```

```python
new_data_set = new_data_set.drop(['id','alco','height','gender'],axis=1)
```

```python
new_data_set
```

```python
x=new_data_set.iloc[:,:-1]
```

```python
x
```

```python
y=new_data_set.iloc[:,8]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.30,random_state=1)
```

```python
xtrain
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rclf = RandomForestClassifier()
```

```python
rclf.fit(xtrain,ytrain)
```

```python
rclf.score(xtest,ytest)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dclf = DecisionTreeClassifier()
```

```python
dclf.fit(xtrain,ytrain)
```

```python
dclf.score(xtest,ytest)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nclf = GaussianNB()
```

```python
nclf.fit(xtrain,ytrain)
```

```python
nclf.score(xtest,ytest)
```

```python
from sklearn.svm import SVC
```

```python
sclf = SVC()
```

```python
sclf.fit(xtrain,ytrain)
```

```python
sclf.score(xtest,ytest)
```
